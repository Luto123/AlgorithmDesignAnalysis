**问题2的Idea：**

***算法思路：*** *原问题可以转换为对于图$G$，是否存在一棵生成树，树中最大边权值不超过$b$，也就是所有的边的权值都不超过$b$。那么有生成树的定义，生成树只能出现在连通图中，那么可以将原来的图G中，边权值大于$b$的所有边移除，判断剩下的边能否将所有的节点连接起来并构成一个连通图，只要是一个连通图，那么就一定存在最小生成树，并且其最大边权值不超过$b$，且由于最小生成树一定是瓶颈生成树，所以就存在瓶颈生成树。*

**问题3的Idea：**

***算法思路：*** *通过二分边权值来确定最小的瓶颈边权，利用问题2中的判定函数逐步逼近最优瓶颈值。找到最小可行瓶颈边权后，我们再次遍历边权不超过该值的边并使用并查集构造出一棵实际的生成树。总共有至多 $E$ 个不同的边权值，对它们进行排序时间为 $O(E \log E)$。二分查找过程最多进行 $O(\log E)$ 次，每次使用线性算法判断可行性，总共 $O(E \log E)$。构造生成树最多遍历一次边，时间为 $O(E)$。所以总时间复杂度为 $O(E \log E)$，接近线性时间复杂度。（assist by GPT.）。*

```python
# 问题2 
# add your code here (assist by GPT.)
def existsBottleneckTree(V, E, w, b):
    """使用并查集的方式去除图中超过阈值的边"""
    parent = list(range(V))
    def find(u):
        while parent[u] != u:
            parent[u] = parent[parent[u]]
            u = parent[u]
        return u

    def union(u, v):
        pu, pv = find(u), find(v)
        if pu != pv:
            parent[pu] = pv
            return True
        return False

    # 筛选所有边权 <= b 的边
    filtered_edges = [(u, v) for (u, v), weight in zip(E, w) if weight <= b]

    # 尝试用这些边构造生成树
    edge_count = 0
    for u, v in filtered_edges:
        if union(u, v):
            edge_count += 1

    # 判断是否构成连通图（生成树应有 V-1 条边）
    return edge_count == V - 1


if __name__ == "__main__":
    V = 4
    E = [(0, 1), (0, 2), (1, 2), (1, 3)]
    w = [1, 4, 2, 3]
    b = 3

    print(existsBottleneckTree(V, E, w, b))
# algorithm of the liear time complexity 
```

```python
# 问题3 
# add your code here (assist by GPT.)
def existsBottleneckTree(V, E, w, b):
    """使用并查集的方式去除图中超过阈值的边"""
    parent = list(range(V))

    def find(u):
        while parent[u] != u:
            parent[u] = parent[parent[u]]
            u = parent[u]
        return u

    def union(u, v):
        pu, pv = find(u), find(v)
        if pu != pv:
            parent[pu] = pv
            return True
        return False

    # 筛选所有边权 <= b 的边
    filtered_edges = [(u, v) for (u, v), weight in zip(E, w) if weight <= b]

    # 尝试用这些边构造生成树
    edge_count = 0
    for u, v in filtered_edges:
        if union(u, v):
            edge_count += 1

    # 判断是否构成连通图（生成树应有 V-1 条边）
    return edge_count == V - 1


def findBottleneckSpanningTree(V, E, w):
    edges = list(zip(E, w))
    unique_weights = sorted(set(w))

    # 使用判定函数进行二分搜索
    def is_feasible(b):
        return existsBottleneckTree(V, E, w, b)

    low, high = 0, len(unique_weights) - 1
    best_bottleneck = None

    while low <= high:
        mid = (low + high) // 2
        b = unique_weights[mid]
        if is_feasible(b):
            best_bottleneck = b
            high = mid - 1
        else:
            low = mid + 1

    # 构造该瓶颈生成树
    parent = list(range(V))

    def find(u):
        while parent[u] != u:
            parent[u] = parent[parent[u]]
            u = parent[u]
        return u

    def union(u, v):
        pu, pv = find(u), find(v)
        if pu != pv:
            parent[pu] = pv
            return True
        return False

    # 构造树
    bst_edges = []
    for (u, v), weight in sorted(edges, key=lambda x: x[1]):
        if weight <= best_bottleneck and union(u, v):
            bst_edges.append((u, v, weight))
            if len(bst_edges) == V - 1:
                break

    return bst_edges, best_bottleneck


if __name__ == "__main__":
    V = 4
    E = [(0, 1), (0, 2), (1, 2), (1, 3)]
    w = [1, 4, 2, 3]

    bst_edges, bottleneck = findBottleneckSpanningTree(V, E, w)
    print(f"瓶颈最小生成树最大边权值为: {bottleneck}")
    print("生成树边列表:")
    for u, v, weight in bst_edges:
        print(f"{u} - {v} (w: {weight})")
# algorithm of the liear time complexity 
```

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


idea:

***算法思路：*** *原问题可以表述为在给定的图$G(V,E,l)$和一组侯选边$E'$，从$E'$中选出一条边，使得给定的两个节点$ s, t \in V $之间缩短的距离最大。那么可以首先使用最短路径算法得到从$s$出发到所有其他点之间的最短距离（当然包含$t$）。然后同样的再使用最短路径算法得出除$t$外所有点到以$t$为终点的最短路径及距离。之后遍历$E'$中的每一条边，考虑如果加入这条边，得到新的$s$到$t$的路径，取得最小值，与原始路径做比较，记录下最大的减少量以及路径，最后得到最优的边。由于最短路径算法在非负边权图上能找到最短路径，且我们枚举所有候选边，因此不会漏解。通过预先计算$s$到$u$和$v$到$t$的最短路径，可以在线性时间内评估每个候选边带来的路径减少量，由此说明算法思路是正确的。*

***时间复杂度分析：*** *实现时，最短路径算法使用Dijkstra算法，所以时间复杂度为$O((|V|+|E|) \log |V|)$，遍历$E'$中每条边时，时间复杂度为$O(|E'|)$，所以总的时间复杂度为$O((|V|+|E| \log |V|+|E'|))$。*

```python
# (The implementation is assisted by GPT, but the idea is my own):
import heapq
from collections import defaultdict


def dijkstra(startNode, graph, vertexCount):
    """使用Dijkstra算法计算从起点startNode到所有顶点的最短路径。"""
    dist = [float('inf')] * vertexCount
    dist[startNode] = 0
    minHeap = [(0, startNode)]
    while minHeap:
        currentDist, currentNode = heapq.heappop(minHeap)
        if currentDist > dist[currentNode]:
            continue
        for neighbor, weight in graph[currentNode]:
            if dist[neighbor] > currentDist + weight:
                dist[neighbor] = currentDist + weight
                heapq.heappush(minHeap, (dist[neighbor], neighbor))
    return dist


def bestNewRoad(vertexCount, edgeList, edgeLengths, candidateEdges, candidateLengths, source, target):
    """寻找加入一条新道路后，从 source 到 target 最短路径减少最多的候选边。"""
    # 构建原图和反向图
    graph = defaultdict(list)
    reverseGraph = defaultdict(list)
    for (u, v), length in zip(edgeList, edgeLengths):
        graph[u].append((v, length))
        reverseGraph[v].append((u, length))
    # 从 source 出发的最短路径
    distFromSource = dijkstra(startNode=source, graph=graph, vertexCount=vertexCount)
    # 从 target 出发的最短路径（反图上 Dijkstra）
    distToTarget = dijkstra(startNode=target, graph=reverseGraph, vertexCount=vertexCount)
    originalDistance = distFromSource[target]
    maxImprovement = 0
    bestEdge = None
    # 遍历所有候选边
    for (u, v), length in zip(candidateEdges, candidateLengths):
        if distFromSource[u] != float('inf') and distToTarget[v] != float('inf'):
            newDistance = distFromSource[u] + length + distToTarget[v]
            improvement = originalDistance - newDistance
            if improvement > maxImprovement:
                maxImprovement = improvement
                bestEdge = (u, v, length)
    return bestEdge, maxImprovement


if __name__ == '__main__':
    vertexCount = 5  # 节点数量
    edgeList = [(0, 1), (1, 2), (2, 4), (0, 3), (3, 4)]  # 边集合
    edgeLengths = [2, 2, 2, 3, 6]  # 边权值
    candidateEdges = [(1, 3), (0, 2), (3, 2)]  # 侯选边
    candidateLengths = [1, 2, 1]  # 候选边长度
    source = 0  # s
    target = 4  # t

    bestEdge, improvement = bestNewRoad(vertexCount=vertexCount, edgeList=edgeList, edgeLengths=edgeLengths,
                                        candidateEdges=candidateEdges, candidateLengths=candidateLengths, source=source,
                                        target=target)

    print(f"最优新增边: {bestEdge}")
    print(f"可减少的最短路径长度: {improvement}")
```

## 问题 4

**逃离问题**

一个 $ n \times n $ 的网格是一个无向图，由 $ n $ 行和 $ n $ 列的顶点组成，如下图所示。我们用 $(i,j)$ 表示第 $ i $ 行和第 $ j $ 列的顶点。除了边界顶点，网格中的所有顶点都有四个邻居，即满足 $ i = 1, i = n, j = 1 $ 或 $ j = n $ 的点 $(i,j)$。

给定网格中的 $ m \leq n^2 $ 个起点 $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$，逃离问题是确定是否存在 $ m $ 条顶点不相交的路径（即路径之间不相交），从这些起点到边界上的任意 $ m $ 个不同点。例如，图1中的网格存在逃离。

(1) 该问题可以看作是一个最大流问题。考虑一个流网络，其中顶点和边都有容量。也就是说，进入任何给定顶点的总正流量受到容量限制。证明在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。更准确地说，你需要将一个具有顶点和边容量的网络 $ G = (V,E) $ 转换为另一个仅具有边容量的网络 $ G' = (V', E') $，使得两个网络上的最大流相同，并且你构建的新网络具有 $ V' = O(V) $ 个顶点和 $ E' = O(E) $ 条边。你可以假设网络是连通的。

(2) 描述一个解决逃离问题的高效算法，并分析其运行时间。

idea：

**问题1：**

*在带有顶点容量和边容量的的网络中，可以通过顶点拆分的方式将顶点容量转换为边容量，将问题转换为只含有边容量的标准的最大流问题。顶点拆分的具体方式为，假设点$A$的容量为$C_A$，那么将点$A$分裂成两个点$A_1$和$A_2$，从点$A_1$到$A_2$的容量为$C_A$，而$A_1$和$A_2$不再具有容量，如果一开始有从点$A$到点$B$路径，且这条路径的容量为$C_{AB}$，那么将点$B$分裂成两个点$B_1$和$B_2$，从点$B_1$到$B_2$的容量为$C_B$，之后处理原先从点$A$到点$B$路径时则设置从点$A_2$到点$B_1$的路径容量为$C_{AB}$。这样就能保证每个点的最大通过流量等于其顶点容量。通过以上操作，点的容量限制就变成了一条边的容量限制。由此证明了在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。*

**问题2：**

*(assist by GPT.)*

*由问题1的提示，逃离问题实际上可以看作寻找m条顶点不相交路径问题，也就是每个顶点的容量为1，可以将其转化为最大流问题。通过问题1的思路进行顶点拆分，拆分后的边的容量为1，此时将逃离问题转换为最大流问题。由于题目中提到的多个起点，那么人为地添加一个虚拟的源点s，然后从s向所有这些起点连接一条容量相应的边。这样，所有的流量就统一地从s发出，便于后续应用标准的最大流算法，且网络中存在多个终点（边界点）可以类似的添加一个虚拟终点t，所有的流量就集中到这个统一的终点，能够更好的判断总共多少条流通到达目标。通过以上过程，这个网络就已经完成了重建，之后直接使用最大流算法。如果最大流等于起点个数m，说明存在m条不重合的逃脱路径，则原问题存在这样的解。*

*在使用顶点拆分等网络重建操作时，需要对每个点进行拆分和连接，总的时间复杂度为$O(n^2)$，重建后网络中点的数量最多为$O(2n^2+2)$，边的数量为$O(n^2)$，常规的最大流求解算法时间复杂度为$O(n^3)$，适用在小规模图上求解，如果图比较复杂，则对应的运行时间呈现指数上升。*
